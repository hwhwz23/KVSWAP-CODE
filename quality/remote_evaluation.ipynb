### Remote Accuracy Evaluation

This notebook is prepared specifically for AE reviewers to run a quick accuracy evaluation **remotely** on our pre-configured server.
The goal is to enable reviewers to perform the evaluation directly in a web browser, **without any setup**.

If you are an AE reviewer and would like to use this remote evaluation workflow, please contact us via HotCRP. We will provide the remote access IP address and credentials promptly.

Please note that, at this stage, we provide one server with 2x RTX 4090 GPUs for AE remote access. Due to time and GPU memory constraints, only a selected subset of experiments and benchmark datasets is run on this machine.
Because of this approximation and the inherent randomness of the experiments,
the reported numbers may differ from those in the paper. Nevertheless, the overall trends should remain consistent, which is sufficient to validate the correctness of the artifact.

If you would like to run the full set of experiments, please follow `README.md` to configure and run the scripts on an A100 (or equivalent) server.



### 1. Basic Setup
Before running the evaluation, please set the two variables: `DS_API_KEY` and `EVAL_USER`.

- `DS_API_KEY`: Some evaluations (Figure-9 and Table-3-Right) require a judge LLM to score model outputs. By default, we use `deepseek-chat` as the judge model, so please provide a valid DeepSeek API key. **If you are an AE reviewer, we can provide an API key upon request via HotCRP.**

- `EVAL_USER` is a short label that separates output folders when multiple people evaluate on the same machine.

- **For remote AE reviewers: choose a unique `EVAL_USER` so your results are not mixed with anyone else's.**


```python
# Set the API key used by judge-LLM evaluations.
DS_API_KEY="your_deepseek_api_key"
# EVAL_USER is a short label that separates output folders when multiple people evaluate on the same machine.
# For remote AE reviewers: choose a unique EVAL_USER so your results are not mixed with anyone else's.
EVAL_USER = 'test0' # **change this to your own unique label**
```
